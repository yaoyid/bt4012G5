```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 500)
```

```python
# Load the data and check the first few lines
dataset_path = "preprocessed_data.csv" # use your own path

dataset = pd.read_csv(dataset_path)
dataset.head()
```

```python
# first create a new column of missing company profile
# 0 - missing, 1- present
dataset['missing company profile'] = dataset['company_profile'].apply(lambda x: 0 if pd.isna(x) else 1)
```

```python
# Then create a new column of missing company profile and company logo
# has_company_logo: 1 - present, 0 - missing
# 0: both present, 1: logo is present, 2: profile is present, 3: both are present
dataset['missing company information'] = dataset['has_company_logo'] + dataset['missing company profile'] * 2
```
